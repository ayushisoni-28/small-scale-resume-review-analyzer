```python
import pdfplumber
import nltk
from collections import Counter
import re

# Download stopwords (you can choose to use them for further filtering if needed)
nltk.download('stopwords')

# List of skills (you can expand this list as needed)
skills_list = ['python', 'java', 'sql', 'machine learning', 'deep learning', 'tensorflow', 'pandas', 'data analysis',
               'data visualization', 'excel', 'r', 'hadoop', 'Exploratory Data Analysis', 'javaScript', 'html', 'css', 'aws',
               'docker', 'EDA', 'etl', 'nosql', 'predictive analytics', 'bi tools', 'big data', 'statistics', 'algorithms',
               'cloud computing', 'nodejs', 'linux', 'git', 'scala', 'matlab', 'cybersecurity', 'tableau', 'powerbi']

# Sample skills required by a business analytics company
business_analytics_skills = ['sql', 'data analysis', 'machine learning', 'data visualization']

# Function to extract text from a PDF resume
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Function to extract skills from the text
def extract_skills_from_text(text, skills_list):
    # Preprocess text: convert to lowercase and remove non-alphabetic characters (punctuation)
    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Initialize a counter to store skill occurrences
    skill_counts = Counter()

    # Check for multi-word skills/phrases first
    for skill in skills_list:
        # Regex to find the skill as a whole phrase (case insensitive)
        pattern = r'\b' + re.escape(skill) + r'\b'
        matches = re.findall(pattern, text)
        skill_counts[skill] += len(matches)

    return skill_counts

# Main function to run the whole process
def main():
    # Assuming the uploaded file is a PDF
    pdf_path = list(uploaded.keys())[0]  # Get the filename of the uploaded PDF
    text = extract_text_from_pdf(pdf_path)  # Extract text from the PDF
    skills_found = extract_skills_from_text(text, skills_list)  # Extract skills from the text

    # Display the extracted skills and their counts (only those with count > 0)
    if skills_found:
        print("Skills found in the resume:")
        for skill, count in skills_found.items():
            if count > 0:  # Show only skills with at least one occurrence
                print(f"{skill.capitalize()}: {count} occurrence(s)")

        # Match skills against business analytics required skills
        matched_skills = [skill for skill in business_analytics_skills if skill in skills_found and skills_found[skill] > 0]

        if matched_skills:
            print(f"\nMatched skills with Business Analytics company (Total: {len(matched_skills)}):")
            print(", ".join(matched_skills))
        else:
            print("\nNo matched skills with Business Analytics company.")
    else:
        print("No relevant skills found in the resume.")

# Run the main function
main()
```
